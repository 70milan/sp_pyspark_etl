```python
import boto3
import json

# Load AWS credentials
secrets_path = "/Workspace/etl_spotify/secrets.json"
with open(secrets_path, "r") as f:
   secrets = json.load(f)

aws_credentials = secrets["aws"]
aws_access_key = aws_credentials["aws_access_key_id"]
aws_secret_key = aws_credentials["aws_secret_access_key"]
s3_bucket = aws_credentials["bucket_name"]

# Create S3 client
s3_client = boto3.client(
   's3',
   aws_access_key_id=aws_access_key,
   aws_secret_access_key=aws_secret_key
)
# ----------------------------------------
# Download and save all_items.json
# ----------------------------------------
response_all_items = s3_client.get_object(Bucket=s3_bucket, Key='all_items.json')
all_items_content = response_all_items['Body'].read().decode('utf-8')

volume_path_all_items = "/Volumes/spotify_pipeline/spotify/raw/all_items.json"
with open(volume_path_all_items, 'w') as f:
   f.write(all_items_content)

print(f"✅ Saved all_items.json to {volume_path_all_items}")

# ----------------------------------------
# Download and save genre_by_artists.json
# ----------------------------------------
response_genres = s3_client.get_object(Bucket=s3_bucket, Key='genre_by_artists.json')  # key name still says .csv
genre_content = response_genres['Body'].read().decode('utf-8')

volume_path_genre_by_artists = "/Volumes/spotify_pipeline/spotify/raw/genre_by_artists.json"  # saving as .json
with open(volume_path_genre_by_artists, 'w') as f:
    f.write(genre_content)

print(f"✅ Saved genre_by_artists.json to {volume_path_genre_by_artists}")

# ----------------------------------------
#                 Parsing
# ----------------------------------------

with open(volume_path_all_items, 'r') as f:
   all_items = json.load(f)

print(f" Number of items in all_items: {len(all_items)}")
print(f" Type of all_items: {type(all_items)}")

with open(volume_path_genre_by_artists, 'r') as f:
    genre_by_artists = json.load(f)

print(f" Number of items in genre_by_artists: {len(genre_by_artists)}")
print(f" Type of genre_by_artists: {type(genre_by_artists)}")
```

```python
# download and save tract_features #

response_tracks_features = s3_client.get_object(Bucket=s3_bucket, Key='tracks_features.csv')  # key name still says .csv
tracks_features = response_tracks_features['Body'].read().decode('utf-8')

volume_path_tracks_features = "/Volumes/spotify_pipeline/spotify/raw/tracks_features.csv"  # saving as .json
with open(volume_path_tracks_features, 'w') as f:
    f.write(tracks_features)

print(f"✅ Saved tracks_features.csv to {volume_path_tracks_features}")
```
